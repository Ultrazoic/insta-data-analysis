```python
#reading the data 
with open("finaldata.txt",encoding="utf-8") as f:
    data=f.read()
```

```python
#spliting data for each individuals
chunks=data.split("\n\n")
chunks=[c for c in chunks if len(c)>3]
```

```python
#parsing data 
def parse_data(chunk):
    try:
        chunk=chunk.strip()
        sep_chunk=chunk.split("\n")
        username=sep_chunk[0]
        posts=int(sep_chunk[1].split(" post")[0].replace(",",""))
        followers=float(sep_chunk[2].split(" follower")[0].replace(",","").replace("K","").replace("M",""))
        if ("K" in sep_chunk[2]):
            followers=int(followers *1000)
        elif ("M" in sep_chunk[2]):
            followers=int(followers *1000000)
        else:
            followers=int(followers)
            
        
        following=float(sep_chunk[3].split(" following")[0].replace(",","").replace("K","").replace("M",""))
        if ("K" in sep_chunk[3]):
            following=int(following *1000)
        elif ("M" in sep_chunk[3]):
            following=int(following *1000000)
        else:
            following=int(following)
            
        name=sep_chunk[4]
        if(len(sep_chunk)>5):
            type_of_page=sep_chunk[5]
            bio="\n".join(sep_chunk[6:])
        else:
            type_of_page="unknown"
            bio=""
        return({"Username":username, "No_of_posts":posts, "No_of_Followers":followers, "No_of_Following":following,
            "Name":name,"Type_of_page":type_of_page,"Bio":bio})
    except Exception as e:
        print(e,chunk)
all_chunks=[]
for c in chunks:
    parsed_data=parse_data(c)
    all_chunks.append(parsed_data)


```

```python
#dumping data and creating a json file
import json
s=json.dumps(all_chunks, indent=4)
with open("data.json" ,"w" , encoding="utf-8") as f:
    f.write(s)
```

```python
# who has max no. of posts
max=0
for chunk in all_chunks:
    if(max<chunk['No_of_posts']):
        max=chunk['No_of_posts']
        chunk_with_maxposts=chunk

print("Max posts - ",chunk_with_maxposts['Username'])
```

```python
#who follows max peoples
max=0
for chunk in all_chunks:
    if(max<chunk['No_of_Following']):
        max=chunk['No_of_Following']
        chunk_with_maxfollowing=chunk
print("Max Following - ",chunk_with_maxfollowing['Username'])
```

```python
# Who has max followers
max=0
for chunk in all_chunks:
    if(max<chunk['No_of_Followers']):
        max=chunk['No_of_Followers']
        chunk_with_maxfollowers=chunk
print("Max Followers - ",chunk_with_maxfollowers['Username'])
```

```python
#Total no. of categories
categories=set()
for chunk in all_chunks:
    categories.add(chunk['Type_of_page'])
print(f"There are total {len(categories)} categories")
```
